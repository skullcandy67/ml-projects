#Introduction:

This dataset contains molecular properties such as MolLogP (lipophilicity), MolWt (molecular weight), NumRotatableBonds (flexibility), and AromaticProportion (aromatic atom ratio). The goal is to predict logS, which represents the aqueous solubility of molecules—a key factor in drug formulation and bioavailability.

We begin by importing the dataset using:
df = pd.read_csv('delaney_solubility_with_descriptors.csv')

This dataset will be explored, processed, and used to train a predictive model for logS.

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer,RobustScaler,PowerTransformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer,r2_score,mean_absolute_error,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import pickle
```

Train-Test Split & Column Transformation:

We begin by splitting the dataset into training and testing sets to evaluate the model on unseen data. To ensure better feature scaling and distribution, we apply the following transformations:

Robust Scaler – Handles outliers by scaling features based on the median and interquartile range.

Power Transformer (Yeo-Johnson) – Corrects skewness, including zero and negative values, making the data more normally distributed for better model performance.

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['logS']), df['logS'], test_size=0.2, random_state=0)
X_train.shape,X_test.shape
```

```python
preprocessor = ColumnTransformer([
    ('MolLogP_new',RobustScaler(),['MolLogP']),
    ('MolWt_new',Pipeline([
        ('MolWt_log',PowerTransformer()),
        ('MolWt_robust',RobustScaler())
         ]),['MolWt','NumRotatableBonds']),
    ('AromaticProportion_new',PowerTransformer(),['AromaticProportion'])
])
```

```python
pipe_lr = Pipeline([
    ('preprocessor',preprocessor),
     ('modelLr',LinearRegression()),
])
pipe_lr.fit(X_train,y_train)
y_pred = pipe_lr.predict(X_test)
print('Linear Regression Score:',r2_score(y_test,y_pred))
print("Linear Regression MSE:", mean_squared_error(y_test, y_pred))
print("Linear Regression MAE:", mean_absolute_error(y_test, y_pred))
```

```python
dt = DecisionTreeRegressor()
param_grid = {
    'max_depth': [3, 5, 7], 
    'min_samples_split': [2, 5, 10],  
    'min_samples_leaf': [2, 4, 6]  
}
scoring = {
    'R2' : 'r2',
    'MSE' : make_scorer(mean_squared_error,greater_is_better=False),
    'MAE' : make_scorer(mean_absolute_error,greater_is_better=False)
}
clf = GridSearchCV(estimator=dt, param_grid=param_grid, scoring=scoring, n_jobs=-1, refit='R2', cv=5)
```

```python
clf.fit(X_train,y_train)
best_dt = clf.best_estimator_
y_pred = best_dt.predict(X_test)
```

```python
print('R2:',r2_score(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print('MAE:',mean_absolute_error(y_test,y_pred))
```

```python
r2 = r2_score(y_test, y_pred)
n = len(y_test)  # Number of samples
k = X_test.shape[1]  # Number of predictors

# Calculate Adjusted R²
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
print('Adjusted R2:',adjusted_r2)
```

```python
print('train_r2',clf.score(X_train, y_train))
print('test_r2',clf.score(X_test, y_test))
```

```python
rf = clf = RandomForestRegressor(random_state=50)
param_grid = {
    'max_depth':[3,5,7],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[2,4,6]
}
scoring = {
    'R2':'r2',
    'MAE':make_scorer(mean_absolute_error,greater_is_better=False),
    'MSE':make_scorer(mean_squared_error,greater_is_better=False),
}
gs = GridSearchCV(estimator=rf, param_grid=param_grid,scoring=scoring, n_jobs=-1, refit='R2', cv=5)
```

```python
gs.fit(X_train,y_train)
gs_better = gs.best_estimator_
y_pred = gs_better.predict(X_test)
```

```python
print('R2:',r2_score(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print('MAE:',mean_squared_error(y_test,y_pred))
```

```python
r2=r2_score(y_test,y_pred)
n=len(y_test)
k=X_test.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print('Adjusted r2:',adjusted_r2)
```

```python
print('Train R2:',gs.score(X_train,y_train))
print('Test R2:',gs.score(X_test,y_test))
```

```python
residual = y_test - y_pred
```
